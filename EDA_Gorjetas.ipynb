##Teste de hipoteses 

**H<sup>null</sup>**
>**A distribuicao da taxa da gorjeta a mesma nos dois grupos**

**H<sup>alt</sup>**
>**A distribuicao da taxa da gorjeta nao a mesma nos dois grupos**

```python
from scipy.stats import ranksums
```

```python
sobremesa = gorjetas.query("sobremesa == 'Sim'").porcentagem
```

```python
sem_sobremesa = gorjetas.query("sobremesa == 'Não'").porcentagem
```

```python
r = ranksums(sobremesa, sem_sobremesa)
```

```python
print('o valor do pvalue e {}'.format(r.pvalue))
```

**H<sup>null</sup>**
>**A distribuicao da taxa da gorjeta a mesma nos dois grupos**

### Teste de hipótese

**H<sup>null</sup>**
>A distribuição do valor da conta é igual no sábado e no domingo

>A distribuição do valor da conta não é igual no sábado e no domingo

```python
r2 = ranksums(valor_conta_sabado, valor_conta_domingo)
print('O valor do p_value é {}'.format(r2.pvalue))
```
